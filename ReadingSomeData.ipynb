## Ploting Data

```python
import matplotlib.pyplot as plt
```

```python
def my_plot():
    df = pd.read_csv("data/AAPL.csv")
    adj_close = df["Adj Close"]
    print(adj_close)
    adj_close.plot()
    plt.show()
#
my_plot()
```

```python
def plot_two_columns():
    df = pd.read_csv("data/AAPL.csv")
    df[["Close", "Adj Close"]].plot()
    plt.show()
#
plot_two_columns()
```
